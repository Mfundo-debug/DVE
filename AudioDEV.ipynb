# 2. Melspectrograms of 3 audio files

```python
import librosa
import librosa.display
```

```python
# set parameters 
winlen_ms = 90 
overlap = 0.2 
sampling_rate = 20055 # use a lower sampling rate 
num_mels = 32

# get the window length in samples 
win_len = int((winlen_ms * sampling_rate) / 1000) 
hop_len = int(win_len * (1 - overlap)) 
nfft = int(2 ** np.ceil(np.log2(win_len)))
```

```python
aud_1, sr1 = librosa.load('0a9f9af7_nohash_0.wav')
aud_2, sr2 = librosa.load('0ac15fe9_nohash_3.wav')
aud_3, sr3 = librosa.load('00b01445_nohash_2.wav')
```

```python
a_1 = librosa.feature.melspectrogram(aud_1, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
a1_DB = librosa.power_to_db(a_1, ref=np.max)
librosa.display.specshow(a1_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Audio 1');
```

```python
a2_S = librosa.feature.melspectrogram(aud_2, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
a2_DB = librosa.power_to_db(a2_S, ref=np.max)
librosa.display.specshow(a2_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Audio 2');
```

```python
a3_S = librosa.feature.melspectrogram(aud_3, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
a3_DB = librosa.power_to_db(a3_S, ref=np.max)
librosa.display.specshow(a3_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Audio 3');
```

# iv. Displaying Spectral Centroid

## Audio 1

```python
import sklearn
spec_centroid = librosa.feature.spectral_centroid(aud_1, sr=sr1)[0]


```

```python
# Computing the time variable for visualization
plt.figure(figsize=(12, 5))
frames = range(len(spec_centroid))
tm = librosa.frames_to_time(frames)
# Normalising the spectral centroid for visualisation
def normal(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
#Plotting the Spectral Centroid along the waveform
librosa.display.waveplot(aud_1, sr=sr1, alpha=0.5)
plt.plot(tm, normal(spec_centroid), color='b');
plt.title('Spectral Centroid and waveform of audio 1');
```

## Audio 2

```python
spec_c2 = librosa.feature.spectral_centroid(aud_2, sr=sr2)[0]
# Computing the time variable for visualization
plt.figure(figsize=(12, 5))
frames2 = range(len(spec_c2))
tm = librosa.frames_to_time(frames2)
# Normalising the spectral centroid for visualisation
def normal(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
#Plotting the Spectral Centroid along the waveform
librosa.display.waveplot(aud_2, sr=sr2, alpha=0.5)
plt.plot(tm, normal(spec_c2), color='r');
plt.title('Spectral Centroid and waveform of audio 2');
```
