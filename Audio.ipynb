```python
import librosa
import librosa.display
```

```python
# set parameters 
winlen_ms = 120 
overlap = 0.5 
sampling_rate = 22050 # use a lower sampling rate 
num_mels = 32

# get the window length in samples 
win_len = int((winlen_ms * sampling_rate) / 1000) 
hop_len = int(win_len * (1 - overlap)) 
nfft = int(2 ** np.ceil(np.log2(win_len)))
```

```python
truck_ms, sr = librosa.load('2019-10-22-16-35_Fraunhofer-IDMT_30Kmh_1620120_A_D_TR_SE_CH34.mp3')
background_ms, sr = librosa.load('2019-10-22-15-30_Fraunhofer-IDMT_30Kmh_8390400_ME_CH12-BG.mp3')
bus_ms, sr = librosa.load('2019-10-22-08-40_Fraunhofer-IDMT_30Kmh_1272258_M_D_BR_ME_CH12.mp3')
car_ms, sr = librosa.load('2019-10-22-08-40_Fraunhofer-IDMT_30Kmh_63533_M_D_CL_ME_CH12.mp3')
motorcycle_ms, sr = librosa.load('2019-10-22-08-40_Fraunhofer-IDMT_30Kmh_461186_M_D_MR_SE_CH34.mp3')
```

```python
truck_S = librosa.feature.melspectrogram(truck_ms, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
truck_S_DB = librosa.power_to_db(truck_S, ref=np.max)
librosa.display.specshow(truck_S_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Truck');
```

```python
background_S = librosa.feature.melspectrogram(background_ms, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
background_S_DB = librosa.power_to_db(background_S, ref=np.max)
librosa.display.specshow(background_S_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Background');
```

```python
car_S = librosa.feature.melspectrogram(car_ms, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
car_S_DB = librosa.power_to_db(car_S, ref=np.max)
librosa.display.specshow(car_S_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Car');
```

```python
bus_S = librosa.feature.melspectrogram(bus_ms, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
bus_S_DB = librosa.power_to_db(bus_S, ref=np.max)
librosa.display.specshow(bus_S_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Bus');
```

```python
motorcycle_S = librosa.feature.melspectrogram(motorcycle_ms, sr=sampling_rate, n_fft=nfft, hop_length=hop_len, n_mels=num_mels, win_length=win_len)
motorcycle_S_DB = librosa.power_to_db(motorcycle_S, ref=np.max)
librosa.display.specshow(motorcycle_S_DB, sr=sampling_rate, x_axis='time', y_axis='mel');
plt.colorbar();
plt.title('Motorcycle');
```
