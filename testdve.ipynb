```python
# BBC dataset
cust_col = ["#1f77b4","#ff7f0e"] # own custom color scheme 
sns.set_palette(cust_col) # set color scheme
total = len(bbc_data)
plt.figure(figsize=(7,5))
g = sns.countplot(x='Label', data=bbc_df)
g.set_ylabel('Count', fontsize=14)
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height + 1.5,
            '{:1.2f}%'.format(height/total*100),
            ha="center", fontsize=14, fontweight='bold')
plt.margins(y=0.1)
plt.title('BBC data Commercial (1) or not (-1)');
plt.xlabel('Commerical');
plt.ylabel('Count');
plt.show()
```

```python
# CNN dataset
cust_col = ["#ff7f0e","#1f77b4"] # own custom color scheme 
sns.set_palette(cust_col) # set color scheme
total = len(cnn_data)
plt.figure(figsize=(7,5))
g = sns.countplot(x='Label', data=cnn_df)
g.set_ylabel('Count', fontsize=14)
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height + 1.5,
            '{:1.2f}%'.format(height/total*100),
            ha="center", fontsize=14, fontweight='bold')
plt.margins(y=0.1)
plt.title('CNN data Commercial (1) or not (-1)');
plt.xlabel('Commerical');
plt.ylabel('Count');
plt.show()
```

# Machine learning

```python
bbc_df['Label'] = bbc_df['Label'].replace(-1,0)
```

```python
bbc_df['Label'].value_counts()
```

```python
import random
random.seed(0)
```

```python
X = bbc_df.iloc[:, : 10]
y = bbc_df.Label
```

```python
for col in X.columns:
    plt.figure()
    sns.distplot(X[col],kde = False)
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X,y)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y, lr_model.predict(X)))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
print(classification_report(y, y_pred))
```

Random forest perfomed better in terms of accuracy (94%) 
and also had  94% for (1) and (-1) for 95% under f1-score
